```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Specify the directory where the CSV files are located
csv_directory = 'C:\\Users\\abdul\\Downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_6298256'

# Get the list of CSV files in the directory
csv_files = [file for file in os.listdir(csv_directory) if file.endswith('.csv')]

# Assume you want to read the first CSV file in the list
if csv_files:
    csv_file_path = os.path.join(csv_directory, csv_files[0])

    # Read the CSV file with adjustments
    df = pd.read_csv(csv_file_path, delimiter=',', skiprows=4, header=None)

    # Extract country names and population data
    country_column = df[0]
    population_column = df.iloc[:, 4:].astype(float).sum(axis=1)  # Sum all population columns

    # Now, you can proceed with the rest of your code to create the bar chart or histogram
    plt.bar(country_column, population_column)
    plt.xlabel('Country')
    plt.ylabel('Population')
    plt.title('Population Distribution by Country')
    plt.xticks(rotation=90)
    plt.show()
else:
    print("No CSV files found in the specified directory.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'df' with relevant columns
# Replace this with your actual DataFrame and column names
# For this example, I'll use a simplified dataset.
data = {
    'Country Name': ['Country1', 'Country1', 'Country2', 'Country2'],
    'Year': [2000, 2010, 2000, 2010],
    'Male Population': [500000, 600000, 700000, 800000],
    'Female Population': [450000, 550000, 650000, 750000]
}
df = pd.DataFrame(data)

# Calculate the male-to-female ratio
df['Male to Female Ratio'] = df['Male Population'] / df['Female Population']

# Pivot the DataFrame to have years as rows and countries as columns
df_pivot = df.pivot(index='Year', columns='Country Name', values='Male to Female Ratio')

# Plotting the line graph for the male-to-female ratio over time
for country in df_pivot.columns:
    plt.plot(df_pivot.index, df_pivot[country], label=country)

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Male to Female Ratio')
plt.title('Male to Female Ratio Over Time for Countries')
plt.legend()
plt.show()
```
